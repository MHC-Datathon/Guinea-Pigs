## Pair Plot Analysis with Another Dataset

Now, let's perform pair plot analysis using another dataset available in the workspace. For demonstration, we will use the ACE violations cleaned dataset again, but you can specify another dataset if needed.

```python
# Example: Load another dataset (can be replaced with any other CSV file)
other_file_path = '/Users/test/MTA-Datathon/dataset-patterns/data/MTA_Bus_Automated_Camera_Enforcement_Violations__Beginning_October_2019_20250922.csv'  # Change this to another dataset if needed
other_df = pd.read_csv(other_file_path)

# Select columns for pairplot (update as needed)
numerical_cols = other_df.select_dtypes(include=['number']).columns.tolist()

# Perform pairplot
sns.pairplot(other_df[numerical_cols])
plt.show()
```
