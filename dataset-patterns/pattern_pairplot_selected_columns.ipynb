# Pair Plot Analysis: ACE Violations Dataset

This notebook will help you explore relationships between selected columns in the main ACE violations dataset using pair plots.

```python
%pip install seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the main ACE violations dataset
file_path = '/Users/test/MTA-Datathon/dataset-patterns/ace_violations_cleaned.csv'
df = pd.read_csv(file_path)

# Display the first few rows and columns for selection
display(df.head())
print('Columns:', df.columns.tolist())
```

## Pair Plot Analysis with MTA Bus Wait Assessment: 2020-2024 Dataset

Now, let's perform pair plot analysis using MTA Bus Wait Assessment: 2020-2024 available dataset in the workspace. 

```python
# Example: Load another dataset (can be replaced with any other CSV file)
other_file_path = 'data/MTA_Bus_Wait_Assessment__2020_-_2024_20250923.csv'
other_df = pd.read_csv(other_file_path)
display(other_df.head())
print('Columns:', other_df.columns.tolist())
# Select columns for pairplot (update as needed)
numerical_cols = other_df.select_dtypes(include=['number']).columns.tolist()

# Perform pairplot
sns.pairplot(other_df[numerical_cols])
plt.show()
```
